```python
eviction_data = 'https://data.sfgov.org/resource/93gi-sfd2.json'

evict_df = pd.read_json(eviction_data)
```

```python
evict_df.columns
```

```python
import numpy as np

evict_df['eviction_reason'] = np.where(evict_df['nuisance'] == 1, 'Nuisance',
                              np.where(evict_df['access_denial'] == 1, 'Access Denial',
                              np.where(evict_df['other_cause'] == 1, 'Other Cause',
                              np.where(evict_df['owner_move_in'] == 1, 'Owner Move In',
                              np.where(evict_df['roommate_same_unit'] == 1, 'Roommate Same Unit',
                              np.where(evict_df['substantial_rehab'] == 1, 'Substantial Rehab',
                              np.where(evict_df['unapproved_subtenant'] == 1, 'Unapproved Subtenant',                                       
                              np.where(evict_df['non_payment'] == 1, 'Non-payment',                                       
                              np.where(evict_df['illegal_use'] == 1, 'Illegal Use',
                              np.where(evict_df['late_payments'] == 1, 'Late Payments',
                              np.where(evict_df['demolition'] == 1, 'Demolition', 
                              np.where(evict_df['breach'] == 1, 'Breach',
                              np.where(evict_df['failure_to_sign_renewal'] == 1, 'Failure to Sign Renewal',                                       
                              np.where(evict_df['capital_improvement'] == 1, 'Capital Improvement',
                              np.where(evict_df['ellis_act_withdrawal'] == 1, 'Ellis Act Withdrawal',
                              np.where(evict_df['development'] == 1, 'Development',                                       
                              np.where(evict_df['good_samaritan_ends'] == 1, 'Good Samaritan Ends',
                              np.where(evict_df['condo_conversion'] == 1, 'Condo Conversion', np.nan))))))))))))))))))
```

```python
cols_to_drop = [':@computed_region_6ezc_tdp2', ':@computed_region_6pnf_4xz7',
       ':@computed_region_9jxd_iqea', ':@computed_region_bh8s_q3mv',
       ':@computed_region_fyvs_ahh9', ':@computed_region_h4ep_8xdi',
       ':@computed_region_p5aj_wyqh', ':@computed_region_rxqg_mtj9',
       ':@computed_region_yftq_j783', 'access_denial', 'breach',
       'capital_improvement', 'condo_conversion', 'demolition', 'development', 'ellis_act_withdrawal', 'failure_to_sign_renewal',
       'good_samaritan_ends', 'illegal_use', 'late_payments',
       'lead_remediation', 'neighborhood', 'non_payment', 'nuisance',
       'other_cause', 'owner_move_in', 'roommate_same_unit',
       'substantial_rehab', 'unapproved_subtenant']

evict_df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
lats = np.array([evict_df['client_location'][0]['coordinates'][0] for coord in evict_df['client_location']])
lons = np.array([evict_df['client_location'][1]['coordinates'][1] for coord in evict_df['client_location']])
```

```python
evict_df['latitude'] = lats
evict_df['longitude'] = lons
```

```python
evict_df.drop('client_location', axis=1, inplace=True)
```
